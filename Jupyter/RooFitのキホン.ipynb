```python
import ROOT as R
```

```python
ws = R.RooWorkspace("ws")
ws.factory("Gaussian::gaus(x[-10,10], mean[0], sigma[1])")
data = ws.obj("gaus").generate( R.RooArgSet(ws.var("x")), 1000)
```

```python
c= R.TCanvas("c","c")

frame = ws.var("x").frame()
data.plotOn(frame)
ws.obj("gaus").plotOn(frame) 

frame.Draw()
c.Draw()
```
